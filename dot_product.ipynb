```python
import nengo
import numpy as np


model = nengo.Network()
with model:
    A = nengo.Node([2, 4])
    B = nengo.Node([3, 1])
    prod = nengo.networks.Product(100, dimensions=2, input_magnitude=5)
    combined = nengo.Ensemble(300, dimensions=1, radius=15)
    
    nengo.Connection(A, prod.A)
    nengo.Connection(B, prod.B)
    nengo.Connection(prod.output, combined, transform=nengo.networks.product.dot_product_transform(2))

    
    c_probe = nengo.Probe(combined, synapse=0.01)
    prod_probe = nengo.Probe(prod.output, synapse=0.01)
```

```python
sim = nengo.Simulator(model)
sim.run(1)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.plot(sim.trange(), sim.data[c_probe])
plt.figure()
plt.plot(sim.trange(), sim.data[c_probe], alpha=0.2, label="approx")
plt.plot(sim.trange(), sim.data[prod_probe], alpha=0.2, label="errrr")
plt.plot(sim.trange(), [np.dot([2,1],[3,2]) for t in sim.trange()], label="actual")
plt.legend()
```
